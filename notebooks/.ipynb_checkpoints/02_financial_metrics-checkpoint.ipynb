# 02 — Financial Metrics

This notebook:
1. Loads raw OHLCV data from `data/raw/` via `src/data_loader.py`
2. Computes financial metrics via `src/metrics.py`
3. Displays a styled summary table
4. Plots a correlation heatmap of daily returns
5. Plots cumulative returns for all tickers
6. Ranks tickers by Sharpe ratio

```python
import sys
import os

# Make src/ importable when the notebook is run from the notebooks/ directory
sys.path.insert(0, os.path.join(os.path.dirname(os.getcwd()), 'src') if os.path.basename(os.getcwd()) == 'notebooks' else os.path.join(os.getcwd(), 'src'))

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

from data_loader import load_data, download_data, DEFAULT_TICKERS
from metrics import compute_metrics

# ── notebook-wide style ──────────────────────────────────────────────────────
sns.set_theme(style='darkgrid', palette='tab10')
plt.rcParams.update({'figure.dpi': 130, 'figure.figsize': (12, 5)})
pd.set_option('display.float_format', '{:.4f}'.format)

print('Environment ready.')
```

## 1 · Load Data

Reads the CSVs that `download_data()` saved in `data/raw/`.  
If a ticker's CSV is missing, `download_data()` is called automatically to fetch it.

```python
data = load_data(DEFAULT_TICKERS)

# If any tickers are missing from disk, download them now
missing = [t for t in DEFAULT_TICKERS if t not in data]
if missing:
    print(f'\nDownloading missing tickers: {missing}')
    download_data(missing)
    data = load_data(DEFAULT_TICKERS)

print(f'\nLoaded tickers : {sorted(data.keys())}')
print(f'Date range     : {next(iter(data.values())).index[0].date()} '
      f'→ {next(iter(data.values())).index[-1].date()}')
```

## 2 · Compute Metrics

```python
summary = compute_metrics(data, benchmark='SPY')
summary
```

## 3 · Styled Summary Table

Colour-codes each column independently: **green → better**, **red → worse**.  
For `max_drawdown` (negative is worse) the gradient is reversed.

```python
# Columns rendered as percentages vs plain numbers
pct_cols  = ['ann_return', 'ann_volatility', 'max_drawdown']
num_cols  = ['sharpe_ratio', 'ma_50', 'ma_200', 'beta', 'last_close']

# Build a display copy with formatted strings
display_df = summary.copy()
for col in pct_cols:
    display_df[col] = summary[col].map('{:.2%}'.format)
for col in num_cols:
    display_df[col] = summary[col].map('{:.2f}'.format)

display_df.columns = [
    'Ann. Return', 'Ann. Volatility', 'Sharpe Ratio',
    'Max Drawdown', 'MA-50', 'MA-200', 'Beta', 'Last Close'
]

# Apply background gradient on the raw numeric summary (not the string display_df)
styled = (
    summary
    .style
    .background_gradient(subset=['ann_return'],     cmap='RdYlGn')
    .background_gradient(subset=['ann_volatility'], cmap='RdYlGn_r')  # lower vol = better
    .background_gradient(subset=['sharpe_ratio'],   cmap='RdYlGn')
    .background_gradient(subset=['max_drawdown'],   cmap='RdYlGn')    # less negative = better
    .background_gradient(subset=['beta'],            cmap='coolwarm')
    .format({
        'ann_return':     '{:.2%}',
        'ann_volatility': '{:.2%}',
        'sharpe_ratio':   '{:.2f}',
        'max_drawdown':   '{:.2%}',
        'ma_50':          '{:.2f}',
        'ma_200':         '{:.2f}',
        'beta':           '{:.2f}',
        'last_close':     '{:.2f}',
    })
    .set_caption('Financial Metrics Summary — all tickers')
    .set_table_styles([{'selector': 'caption',
                        'props': [('font-size', '14px'), ('font-weight', 'bold')]}])
)

styled
```

## 4 · Correlation Heatmap of Daily Returns

Shows how closely each pair of tickers moves together.  
Values near **+1** = highly correlated; near **0** = uncorrelated (good for diversification).

```python
# Build a wide daily-returns matrix (rows = dates, columns = tickers)
returns = pd.DataFrame({
    ticker: df['Close'].pct_change().dropna()
    for ticker, df in data.items()
})

corr = returns.corr()

fig, ax = plt.subplots(figsize=(9, 7))
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)   # hide upper triangle (redundant)

sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn',
    vmin=-1, vmax=1,
    linewidths=0.5,
    square=True,
    ax=ax,
)
ax.set_title('Pairwise Correlation of Daily Returns', fontsize=14, pad=12)
plt.tight_layout()
plt.show()
```

## 5 · Cumulative Returns — All Tickers

Each series is normalised to **1.0 on the first common trading date**, so tickers with
different absolute price levels are directly comparable.
A value of 1.20 means the ticker is up 20 % from the base date.

```python
# Cumulative return = (1 + r_t).cumprod(), starting from 1.0
cum_returns = (1 + returns).cumprod()

# Sort legend by final cumulative return (best performer first)
final_values  = cum_returns.iloc[-1].sort_values(ascending=False)
ordered_tickers = final_values.index.tolist()

fig, ax = plt.subplots(figsize=(13, 6))

colors = plt.cm.tab10.colors
for i, ticker in enumerate(ordered_tickers):
    ax.plot(
        cum_returns.index,
        cum_returns[ticker],
        label=f'{ticker}  ({final_values[ticker]:.2f}×)',
        color=colors[i % len(colors)],
        linewidth=1.6,
    )

ax.axhline(1.0, color='grey', linestyle='--', linewidth=0.8, label='Baseline (1.0)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda y, _: f'{y:.1f}×'))
ax.set_title('Cumulative Returns (normalised to 1.0 at start)', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Growth of $1 invested')
ax.legend(loc='upper left', fontsize=9, framealpha=0.8)
plt.tight_layout()
plt.show()
```

## 6 · Sharpe Ratio Ranking

Higher Sharpe = better **risk-adjusted** return.  
Bars are colour-coded: green ≥ 1.0 (strong), yellow 0–1 (acceptable), red < 0 (poor).

```python
sharpe_ranked = summary['sharpe_ratio'].sort_values(ascending=True)  # ascending for horizontal bar

# Colour each bar based on Sharpe threshold
bar_colors = [
    '#2ecc71' if v >= 1.0 else '#f39c12' if v >= 0 else '#e74c3c'
    for v in sharpe_ranked
]

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.barh(sharpe_ranked.index, sharpe_ranked.values, color=bar_colors, edgecolor='white')

# Annotate each bar with the numeric value
for bar, val in zip(bars, sharpe_ranked.values):
    x_pos = bar.get_width() + (0.02 if val >= 0 else -0.02)
    ha = 'left' if val >= 0 else 'right'
    ax.text(x_pos, bar.get_y() + bar.get_height() / 2,
            f'{val:.2f}', va='center', ha=ha, fontsize=10)

ax.axvline(0, color='grey', linewidth=0.8)
ax.axvline(1, color='#2ecc71', linewidth=1, linestyle='--', alpha=0.6, label='Sharpe = 1.0')
ax.set_title(f'Tickers Ranked by Sharpe Ratio  (rf = 5 %)', fontsize=13)
ax.set_xlabel('Sharpe Ratio')
ax.legend(fontsize=9)
plt.tight_layout()
plt.show()

# Print ranked table
print('\n── Sharpe Ratio Ranking (best → worst) ──')
ranked_table = summary[['sharpe_ratio', 'ann_return', 'ann_volatility', 'max_drawdown']].sort_values(
    'sharpe_ratio', ascending=False
)
ranked_table.columns = ['Sharpe', 'Ann. Return', 'Ann. Vol.', 'Max Drawdown']
print(ranked_table.to_string(float_format=lambda x: f'{x:.4f}'))
```
